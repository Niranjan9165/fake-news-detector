````python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

def clean_text(text):
    # Step 1 - Make everything lowercase
    text = text.lower()
    
    # Step 2 - Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Step 3 - Split into individual words
    words = word_tokenize(text)
    
    # Step 4 - Remove stopwords (common words like "the", "is", "at")
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if w not in stop_words]
    
    # Step 5 - Lemmatize (convert words to base form)
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(w) for w in words]
    
    # Join words back together
    return " ".join(words)

# Test it!
fake_headline = "SHOCKING: Scientists PROVE earth is flat!!!"
real_headline = "Federal Reserve raises interest rates by 0.25%"

print("🔴 FAKE headline:")
print("Before:", fake_headline)
print("After:", clean_text(fake_headline))

print("\n🟢 REAL headline:")
print("Before:", real_headline)
print("After:", clean_text(real_headline))
```

You should see:
```
🔴 FAKE headline:
Before: SHOCKING: Scientists PROVE earth is flat!!!
After: shocking scientist prove earth flat

🟢 REAL headline:
Before: Federal Reserve raises interest rates by 0.25%
After: federal reserve raise interest rate
````
