```python
headlines = [
    "Aliens land in New York",
    "President signs new climate bill",
    "Scientists cure common cold",
    "Local dog wins election"
]

print("First headline:", headlines[0])
print("Last headline:", headlines[-1])
print("Total headlines:", len(headlines))
```

```python
confidence_score = 0.87

if confidence_score > 0.8:
    verdict = "Very likely FAKE"
elif confidence_score > 0.5:
    verdict = "Possibly fake"
else:
    verdict = "Likely REAL"

print("Confidence score:", confidence_score)
print("Verdict:", verdict)
```

```python
def analyze_headline(headline):
    words = headline.split()
    word_count = len(words)
    has_exclamation = "!" in headline
    is_all_caps = headline.isupper()
    
    return {
        "headline": headline,
        "word_count": word_count,
        "has_exclamation": has_exclamation,
        "is_all_caps": is_all_caps
    }

result = analyze_headline("SHOCKING: Scientists PROVE earth is flat!!!")
print(result)
```

```python
def basic_fake_news_screener(headline):
    suspicious_words = [
        "SHOCKING", "BREAKING", "MIRACLE", 
        "SECRET", "CURE", "BANNED", "PROVEN"
    ]
    
    headline_upper = headline.upper()
    found_suspicious = []
    
    for word in suspicious_words:
        if word in headline_upper:
            found_suspicious.append(word)
    
    exclamation_count = headline.count("!")
    
    score = 0
    score += len(found_suspicious) * 20
    score += exclamation_count * 15
    score = min(score, 100)
    
    if score >= 60:
        verdict = "⚠️  LIKELY FAKE"
    elif score >= 30:
        verdict = "🤔 SUSPICIOUS"
    else:
        verdict = "✅ LOOKS REAL"
    
    return {
        "headline": headline,
        "score": score,
        "verdict": verdict,
        "suspicious_words": found_suspicious
    }

test_headlines = [
    "SHOCKING SECRET CURE Doctors Don't Want You to Know!!!",
    "Federal Reserve raises interest rates by 0.25%",
    "MIRACLE weight loss PROVEN by scientists!!!",
    "Local school board votes on new curriculum"
]

print("=" * 50)
print("   FAKE NEWS SCREENER RESULTS")
print("=" * 50)

for headline in test_headlines:
    result = basic_fake_news_screener(headline)
    print(f"\nHeadline: {result['headline']}")
    print(f"Score: {result['score']}/100")
    print(f"Verdict: {result['verdict']}")
    print("-" * 40)
```
